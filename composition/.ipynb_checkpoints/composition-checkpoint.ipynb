# composition


Personal reflections on making music.


## realizations


At some point -- say while taking piano lessons as a kid -- at some point we have this small revelation
that we do not *have* to play what is on the page. I can still remember the single note in *The Popcorn Man* in
John Thompson's Modern Course For The Piano where I made a mistake and liked it better than what was written.
Playing that wrong note was *subversive*.


My view of composing has come to this: There is my history, a complex of thoughts and ideas and tunes; and then
there is a blank page. And then there is the third thing and then there is the fourth thing. 


The third thing is the Golem of Disapprobation: All the forces in the universe that want to stop me from making
music. To paraphrase Samuel Beckett: *Every note is like an unnecessary stain on silence and nothingness.* This is
what the cosmos has to say; and it filters through all sorts of channels to become a kind of fractal pressure to
give up give up give up. It is exhausting just contemplating how many things there are trying to stop us. 


The fourth thing is getting pissed at the third thing and not letting it win. I like the expression "our emotions
tell us what is important" over and above what we get from reason helping us be right about things. 


So I would say the first step in composing is clearing away distractions, taking a good hard look at and listen to
the cosmos telling us we can't get anywere... and then just getting pissed about it and creating music anyway 
out of sheer spite and disdain. I think Frank Zappa would smile at this.


## environment


I will do one of three things to get going. Either I'll set up some sort of percussion track; or I will sit down
at the piano/synth; or I will pick up a guitar. I am as happy writing instrumentals as songs; but lyrics have a
certain charm so they are fun to include them when feasible, despite one perhaps not having anything particular 
to say. In fact I often steal lyrics from playwrights and poets to avoid having to come up with anything original. 
This is cheaping out on having a point for the song; and this in turn is an example of fractal pressure: The use of the term 
'cheaping out' is self-critical and I'm using it ironically to support the practice of -- absolutely no argument -- 
ripping off Shakespeare to get some singing into the tune. In summary, having words to focus on is a great feature 
of a song; so not having anything in particular to talk about should not be an obstacle. Here are two elaborations
on the aspect of composing that binds music to words...


First elaboration: Out In The Country. This is a song recorded by Three Dog Night that has a fantastic intro guitar
vamp that quickly welcomes in the promising lyric "Whenever I need to leave it all behind..." For me the first few lines
of Out In The Country are a great example of how quickly and easily music and singing can sweep us away into an immersive
listening frame of mind. But this particular example -- for me -- absolutely crashes and burns a moment later by pivoting 
into a saccharine chorus with matching vacuous, sentimental lyrics; 'Out where the rivers like to run' and so forth. I 
don't fault the tune as a popular success; but it helps me articulate what I am looking for (not that!) when I'm trying 
to put a piece together.


Implicit in this narrative is the idea of a spectrum: On the one side familiarity and resolution and on the other
in contrast tension and novelty. How the song unfolds on this spectrum is a kind of puzzle to solve; so composing a
piece is a process of solving this problem in some idiomatic context. To belabor this a little: By idiomatic context 
I mean something like Samba or Bluegrass. More broadly a collection of songs can be seen as a metasolution of the same
problem, like on the scale of a traditional album of seven or eight tunes. I would say my personal aversion to the
Out In The Country chorus is a disagreement about pegging the spectrum needle to familiar and resolved.


As a second elaboration: Virtually any Yes song features iceberg lyrics that imply much more unsaid while 
evoking themes and events and emotion. As a result I experience this music as *unresolved*. I mean
I dunno, you tell me: Did we really count to 100? The fact of this unresolution that I carry around for
forty years and more is a different facet of engagement, what I referred to above as immersive listening.


This "puzzle thinking" and "lyrical contribution" are deconstructed fragments of a composing process. I think anything
I try and write about composing is going to come down to a bucket of fragments. Here is another: A particular Hungarian
mathematician was famously both an atheist and an appreciator of beauty in mathematical proofs. The most striking
and elegant of these are written by God in a book called The Book, he would say. And his point was that one does not 
necessarily need to believe in God but one *should* believe in *The Book*. In my view every piece of music we have ever
performed or tried to perform is a momentary decoration of time. We can choose to apply a similar aesthetic in 
appreciation. As a starting point all music has merit, has value. When the song ends and supposing we understand 
time as a decorated structure: We might say that the decoration is not with
us at the moment; but persists. I'd like to imagine it has gone into the wind.


## mechanics


A sequencer is very useful for experimentation from a baseline track, for example a chord sequence or an
ostinato. For one thing the synthesizer never gets tired; and it likely has a bag of voices up its sleeve; so over 
a piano one can try out a horn, a flute, an oboe, and a clarinet. In contrast the problem with recording a live
instrument as part of a composition is that it can represent a lot of work to transcribe. This brings us smack
into 'What is the outcome: Recording, score, or a pleasant sunday afternoon?' 


A recording can be played back. Hit play.


A score can be picked up and reinterpreted by other folks. Hit print. 


A pleasant sunday afternoon leaves everything in the wind.



## who knew that would work


Early on in my history of composition -- about 1990 -- I was working on a piece called "Look Up"
which had two distinct melodic lines. Cramming two or more songs worth of music into a single piece
has been a recurring theme. Anyway at some point I managed to record one theme and as I was playing
it back I found I could play the second theme over it. This helped to demystify the idea that
songs are somehow hermetic creations that spring to life fully formed according to formal and
highly restrictive laws of harmony. (More fractal pressure, in other words.)


This realization that lines can be mixed has a recurring value. A bassline can be thought of as
a counterpoint melody in addition to being a reinforcement of the harmonic basement. 



## jingle bells


Given some piece of scaffolding for a song, let's say a chord sequence, I found that other 
songs or themes can often be stretched across it in an interesting way. Without breaking anything. 
As an example back around 1999 I was working on a Christmas-themed tune called SleiBellz. 
The song directly quotes the familiar (and fairly complicated) Christmas tune "Sleigh Ride". My
composition runs pretty long, about 7 minutes, as it evolves through multiple phases with occasional 
callbacks. At one point in the progression I was playing a phrase on the piano based on A-minor.
Since the song is a wry reflection on Christmas music I thought to try playing Jingle Bells starting
with this A-minor chord; and to my surprise this was perfectly feasible with a little
flex in rhythm and voicing. 


So there is a lot more latitude in putting a piece together because the interpretation happens 
through the mind of the listener. That's where this leads: To respect for the human mind as it
can take in and process and connect and be carried along and be carried away by music.
